### 4.1 튜토리얼 column_chart

```python
import xlsxwriter  
# 막대그래프 그리기 

workbook=xlsxwriter.Workbook('tutorial_4_1.xlsx')  
worksheet = workbook.add_worksheet()
chart = workbook.add_chart({'type': 'column'}) 
bold = workbook.add_format({'bold': True}) 
# 데이터 만들기
chart.set_x_axis({
'name': 'student score',
'name_font': {'size': 14, 'bold': True},
})
data = [
[70, 64, 30],
[45, 33, 90],
[25, 44, 30],
]
worksheet.write(0, 1, 'math', bold)  
worksheet.write(0, 2, 'sciece', bold)  
worksheet.write(0, 3, 'computer', bold)  
worksheet.write(1, 0, 'hojun', bold)  
worksheet.write(2, 0, 'eunjung', bold)  
worksheet.write(3, 0, 'subin', bold)  
worksheet.write(4, 0, 'Average', bold)  
worksheet.write_column('B2', data[0])  
worksheet.write_column('C2', data[1])  
worksheet.write_column('D2', data[2])  

chart.add_series({'categories': '=Sheet1!$A$2:$A$4',  
                  'values': '=Sheet1!$B$2:$B$4',  
                  'name':'Sheet1!$B$1'})

chart.add_series({'values': '=Sheet1!$C$2:$C$4',
                  'name':'Sheet1!$C$1'})

chart.add_series({'values': 
                  '=Sheet1!$D$2:$D$4', 
                  'name':'Sheet1!$D$1'})

worksheet.insert_chart('A7', chart)

workbook.close()
```

### 4.2 튜토리얼 bar_chart

```python
import xlsxwriter  
# 평균을 추가하고 데이터의 막대그래프를 세로로 그리기 

workbook=xlsxwriter.Workbook('tutorial_4_2.xlsx')  
worksheet = workbook.add_worksheet()

chart = workbook.add_chart({'type': 'bar'})  
bold = workbook.add_format({'bold': True}) 

chart.set_x_axis({
'name': 'student score',
'name_font': {'size': 14, 'bold': True},
})

data = [
[70, 64, 30],
[45, 33, 90],
[25, 44, 30],
]
worksheet.write(0, 1, 'math', bold)  
worksheet.write(0, 2, 'sciece', bold)  
worksheet.write(0, 3, 'computer', bold)  
worksheet.write(1, 0, 'hojun', bold)  
worksheet.write(2, 0, 'eunjung', bold)  
worksheet.write(3, 0, 'subin', bold)  
worksheet.write(4, 0, 'Average', bold)  
worksheet.write_column('B2', data[0])  
worksheet.write_column('C2', data[1])  
worksheet.write_column('D2', data[2])  
# 평균 추가하기 
worksheet.write(4, 1, '=AVERAGE(B2:B4)')  
worksheet.write(4, 2, '=AVERAGE(C2:C4)')  
worksheet.write(4, 3, '=ROUND(AVERAGE(B2:B4),0)')
# 평균 추가하기 
chart.add_series({'categories': '=Sheet1!$A$2:$A$5',  
                  'values': '=Sheet1!$B$2:$B$5',  
                  'name':'Sheet1!$B$1'})

chart.add_series({'values': '=Sheet1!$C$2:$C$5',
                  'name':'Sheet1!$C$1'})  
chart.add_series({'values': '=Sheet1!$D$2:$D$5', 
                  'name':'Sheet1!$D$1'})

worksheet.insert_chart('A7', chart)

workbook.close()
```

```python
import pandas as pd
import xlsxwriter

data = {
    '이름' : ['호준', '길동', '준호'],
    '국' : [70, 80, 90],
    '영' : [55, 45, 79],
    '수' : [30, 90, 85]
}

df = pd.DataFrame(data)
df = df.set_index('이름')

writer = pd.ExcelWriter('tutorial_3_4_pandas_data.xlsx', engine='xlsxwriter')

df.to_excel(writer, sheet_name='First')

workbook = writer.book
worksheet = writer.sheets['First']

chart = workbook.add_chart({'type':'line'})

chart.set_x_axis({
    'name' : 'Student Score',
    'name_font': {'size': 14, 'bold':True},
})

bold = workbook.add_format({'bold':True})

cell_format = workbook.add_format()
cell_format.set_align('center')
cell_format.set_bold()


worksheet.write(4, 0, '평균', cell_format)
worksheet.write(4, 1, '=ROUND(AVERAGE(B2:B4), 2)')
worksheet.write(4, 2, '=ROUND(AVERAGE(C2:C4), 2)')
worksheet.write(4, 3, '=ROUND(AVERAGE(D2:D4), 2)')


chart.add_series({
    'categories' : '=First!$A$2:$A$4', #호준, 길동, 준호
    'values' : '=First!$B$2:$B$4', #국어 Value
    'name' : '=First!$B$1' #국어
})

chart.add_series({
    'values' : '=First!$C$2:$C$4', #영어 Value
    'name' : '=First!$C$1' #영어
})

chart.add_series({
    'values' : '=First!$D$2:$D$4', #수학 Value
    'name' : '=First!$D$1' #수학
})

worksheet.insert_chart('A7', chart)

workbook.close()
writer.save()
```

### 4.4 튜토리얼  area_chart

```python
# 위에 데이터를 가지고 area_chart 그리기 

import pandas as pd
import xlsxwriter

data = {'Name' : ['hojun', 'enujung' , 'subin'] ,
                   'Math' : [70, 64, 30], 
                   'Science' : [45, 33, 90],
                   'Computer' :[25, 44, 30]  } 

# 평균을 데이터 프레임으로 구하기
df = pd.DataFrame(data)
df = df.set_index('Name')

# average 라는 행을 추가 하여 각 열의 평균을 구하는 방법
df.loc['average'] = df.mean(axis=0)
df.loc['sum'] = df.sum(axis=0)# 합 구하기 

writer = pd.ExcelWriter('tutorial_4_4.xlsx', engine= 'xlsxwriter')

df.to_excel(writer , sheet_name = 'First')

workbook = writer.book
worksheet = writer.sheets['First']
# 이부분만 바꾸면 됩니다 ------------------
chart = workbook.add_chart({'type': 'area'})  
bold = workbook.add_format({'bold': True})  
#-------------------------------------------

chart.set_x_axis({
'name': 'student score',
'name_font': {'size': 14, 'bold': True},
})

chart.add_series({'categories': '=First!$A$2:$A$5',  
                  'values': '=First!$B$2:$B$5',  
                  'name':'First!$B$1'})

chart.add_series({'values': '=First!$C$2:$C$5',
                  'name':'First!$C$1'})  
chart.add_series({'values': '=First!$D$2:$D$5', 
                  'name':'First!$D$1'})
worksheet.insert_chart('A7', chart)

workbook.close()
writer.save()
```
