### 6.1 python으로 메일보내기 

#### 6.1.1 네이버에서 메일 보내기

```python
# SMTP를 이용해서 메일을 보내봅니다.  
import smtplib
from email.mime.text import MIMEText

# 먼저 네이버에서 SMTP를 설정해주세요.  
NAVER_ID = '' # Naver ID
NAVER_PW = '' # Naver password

My_email = '{id}@naver.com'.format(id=NAVER_ID)

server = smtplib.SMTP_SSL('smtp.naver.com', 465)  
server.login(NAVER_ID, NAVER_PW)

Contents = '''안녕하세요. 파이썬으로 메일을 보내봅니다.'''  


msg = MIMEText(Contents, _charset='euc-kr')  
msg['Subject']= '안녕하세요'
msg['From'] = My_email
msg['To'] = '' # 메일을 보내고 싶은 주소를 적어주세요.

server.sendmail(msg['From'], [msg['To']], msg.as_string())  
server.quit()
```

#### 6.1.2 개인별 성적표 만들기

챕터 4에서 만든 성적표는 세명의 성적이 모두 들어간 성적표였습니다.
이번에는 본인의 성적표와 평균만 들어간 성적표를 엑셀로 만들어보겠습니다

```python
import xlsxwriter

data = [
['hojun', '보내실분 id@naver.com', 70, 64, 30],
['eunjung', '보내실분 id@naver.com', 45, 33, 90],
['subin', '보내실분 id@naver.com', 25, 44, 30],
]

math_average = (data[0][2] + data[1][2]+ data[2][2]) / 3  
science_average = (data[0][3] + data[1][3]+ data[2][3]) / 3  
computer_average = (data[0][4] + data[1][4]+ data[2][4]) / 3

for student in data:
    filename = 'tutorial_6_2_{}.xlsx'.format(student[0]) 
    workbook = xlsxwriter.Workbook(filename)  
    worksheet = workbook.add_worksheet()

    chart = workbook.add_chart({'type': 'radar'}) 
    bold = workbook.add_format({'bold': True})

    chart.set_x_axis({  'name': 'student score',
    'name_font': {'size': 14, 'bold': True},
    })

    worksheet.write(0, 0, 'name', bold)  
    worksheet.write(0, 1, 'math', bold)  
    worksheet.write(0, 2, 'sciece', bold)  
    worksheet.write(0, 3, 'computer', bold)

    worksheet.write(1, 0, student[0], bold)  
    worksheet.write(1, 1, student[2], bold) 
    worksheet.write(1, 2, student[3], bold) 
    worksheet.write(1, 3, student[4], bold)

    worksheet.write(2, 0, 'Average', bold) 
    worksheet.write(2, 1, math_average, bold)
    worksheet.write(2, 2, science_average, bold) 
    worksheet.write(2, 3, computer_average, bold)

    chart.add_series(
    {
    'categories': '=Sheet1!$B$1:$D$1',  'values': '=Sheet1!$B$2:$D$2',  'name':'Sheet1!$A$2'
    }
    )
    chart.add_series(
    {
    'values': '=Sheet1!$B$3:$D$3',  'name':'Sheet1!$A$3'
    }
    )
    worksheet.insert_chart('A7', chart) 
    workbook.close()


```

#### 6.1.3 개인별 성적표 만들고 메일보내기 

```python
# Excel에 있는 목록의 주소에 메일을 보냅니다.  import xlsxwriter
import smtplib
from email.mime.text import MIMEText  
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart  
from email import encoders

NAVER_ID = '' # Naver ID
NAVER_PW = '' # Naver password

My_email = '@naver.com'.format(id=NAVER_ID)

server = smtplib.SMTP_SSL('smtp.naver.com', 465) 
server.login(NAVER_ID, NAVER_PW)

data = [
['hojun', '보내실분 id@naver.com', 70, 64, 30],
['eunjung', '보내실분 id@naver.com', 45, 33, 90],
['subin', '보내실분 id@naver.com', 25, 44, 30],
]

math_average = (data[0][2] + data[1][2]+ data[2][2]) / 3 
science_average = (data[0][3] + data[1][3]+ data[2][3]) / 3
computer_average = (data[0][4] + data[1][4]+ data[2][4]) / 3

for student in data:
# 성적표를 작성합니다.
    filename = 'tutorial_6_2_{}.xlsx'.format(student[0])  
    workbook = xlsxwriter.Workbook(filename) 
    worksheet = workbook.add_worksheet()

    chart = workbook.add_chart({'type': 'radar'})  
    bold = workbook.add_format({'bold': True})

    chart.set_x_axis({  'name': 'student score',
    'name_font': {'size': 14, 'bold': True},
    })


    worksheet.write(0, 0, 'name', bold)  
    worksheet.write(0, 1, 'math', bold)  
    worksheet.write(0, 2, 'sciece', bold) 
    worksheet.write(0, 3, 'computer', bold)

    worksheet.write(1, 0, student[0], bold) 
    worksheet.write(1, 1, student[2], bold) 
    worksheet.write(1, 2, student[3], bold) 
    worksheet.write(1, 3, student[4], bold)

    worksheet.write(2, 0, 'Average', bold)  
    worksheet.write(2, 1, math_average, bold) 
    worksheet.write(2, 2, science_average, bold) 
    worksheet.write(2, 3, computer_average, bold)

    chart.add_series(
    {
    'categories': '=Sheet1!$B$1:$D$1',  'values': '=Sheet1!$B$2:$D$2',  'name':'Sheet1!$A$2'
    }
    )
    chart.add_series(
    {
    'values': '=Sheet1!$B$3:$D$3',  'name':'Sheet1!$A$3'
    }
    )

    worksheet.insert_chart('A7', chart)
    workbook.close()

# 메일을 작성 합니다.  
    mail = MIMEMultipart()
    mail['Subject'] = '성적통지서'  
    mail['From'] = My_email 
    mail['To'] = student[1]

    Contents =   """
    안녕하세요, {name}님  성적표를 송부드립니다.
    """.format(name=student[0])
    msg = MIMEText(Contents, 'plain', _charset='euc-kr')
    mail.attach(msg)

    # 파일 첨부을 첨부합니다.
    excelfile = MIMEBase('application', 'octet-stream')
    excelfile.set_payload(open(filename, 'rb').read()) 
    encoders.encode_base64(excelfile)
    excelfile.add_header(
    'Content-Disposition',  'attachment;filename="{file}"'.format(file=filename))
    mail.attach(excelfile)
    server.sendmail(My_email, [student[1]], mail.as_string())  
server.quit()
```

### 6.2 참가신청서 

앞에서 배운 한글 크롤링을 이용하여 여러 한글 파일에서 필요한 정보만을 빼내어 엑셀로 만드는 작업입니다

```python
# 총 6개의 hwp 파일입니다 
import glob
import xlsxwriter
import numpy as np
import pandas as pd
import ole
path = ('./data2/*.hwp')
workbook = xlsxwriter.Workbook('tutorial_6_2.xlsx')
worksheet = workbook.add_worksheet()
num = 1
# 각 한글 폴더에서 성명 , 소속, 연락처 ,e-mail 만 추출  
df = ['번호', '성    명','소    속','연 락 처', 
                  'E-mail']
for n,i in enumerate(df):
    worksheet.write(0,n ,i)

for fname in glob.glob(path):
       with ole.open(fname) as f:
            data = f.get_stream('PrvText').read().decode('utf-16le')
            data = data.split('><')
            input_data = []
            input_data.append(num)
            for i in range(len(data)):
                for j in range(len(df)):
                    if data[i] == df[j]:
                        input_data.append(data[i+1])
            for number,values in enumerate(input_data):
                worksheet.write(num ,number, values)
            num += 1

workbook.close()

       
```

위에 데이터에서 참가신청서 합격 여부를 입력하면 자동으로 메일을 보내는 시스템을 구축하겠습니다

```python
# 합격여부는 엑셀로 작성하는것도 좋은 방법입니다 
import pandas as pd

data = pd.read_excel('./tutorial_6_2.xlsx')
df = pd.DataFrame(data)
df = df.set_index('번호')
df

writer = pd.ExcelWriter('tutorial_6_3.xlsx', engine = 'xlsxwriter')

df.to_excel(writer, sheet_name = 'sheet')

workbook = writer.book
worksheet = writer.sheets['sheet']
bold = workbook.add_format({'bold': True})

worksheet.write(0,5,'합격여부' , bold)
worksheet.write(1,5,'합')
worksheet.write(2,5,'불')
worksheet.write(3,5,'합')
worksheet.write(4,5,'합')
worksheet.write(5,5,'합')
worksheet.write(6,5,'불')

workbook.close()
```

#### 6.2.1 메일보내기

```python
import smtplib
from email.mime.text import MIMEText  
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart  
from email import encoders

NAVER_ID = '' # Naver ID
NAVER_PW = '' # Naver password

My_email = ''.format(id=NAVER_ID)

server = smtplib.SMTP_SSL('smtp.naver.com', 465) 
server.login(NAVER_ID, NAVER_PW)

data = pd.read_excel('./tutorial_6_3.xlsx')
df = pd.DataFrame(data)

for i in range(len(df)):
    # 만약 합격이면 
    if df.iloc[i:i+1,5].values == '합':
        Contents  =  '''안녕하세요 xlsxwrite 튜토리얼 입니다 ,
        {name}님의 참가신청결과는 합격입니다.'''.format(name =df.iloc[i:i+1, 1].values)
    # 불합격이면 
    else:
        Contents  =  '''안녕하세요 xlsxwrite 튜토리얼 입니다 ,
        {name}님의 참가신청결과는 불합격입니다. 아쉽지만 다음기회에도 도전해주세요!!'''.format(name =df.iloc[i:i+1, 1].values)
    # 위에와 동일한 방법으로 보내는 방법
    msg = MIMEText(Contents, 'plain', _charset='euc-kr')
    msg['result'] = '참가신청 결과'
    msg['From'] = My_email
    send = df.iloc[i:i+1, 4].values
    msg['To'] = send[0]
    
    server.sendmail(msg['From'], [msg['To']], msg.as_string())  
server.quit()
    
    
```
