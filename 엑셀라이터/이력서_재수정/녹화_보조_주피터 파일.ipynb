```python
# 성명, 생년월일, 현주소 , 연락처 , 이메일 
def Privacy(data):
    data_out = []
    df = ['성          명' , '생  년  월  일', '현    주    소', '연    락    처','이    메    일']
    for i in range(len(data)):
        for j in range(len(df)):
            if data[i] == '이    메    일':
                a = data[i+1].find('>')
                data_out.append(data[i+1][:a])
                break
            if data[i] == df[j]:
                data_out.append(data[i+1])
    return data_out

# 학교명 , 전공 , 성적, 졸업구문, 소재지 
# 성적 4.5 기준으로 4.3 이상 10 , 4.3~4.1 8점, 4.1~3.9 6점 ,4.1이하 4점 
def School(data):
    data_out = []
    df = ['학교명','전공','성적','졸업구분','소재지']
    for i in range(len(data)):
        for j in range(len(df)):
            if data[i] == '성적':
                if data[i+5] != '':
                    score = data[i+5]
                    score = score.replace('/4.5','')
                    score = float(score)
                    if score >=4.3 : 
                        data_out.append(10)
                        break
                    elif 4.3>score>=4.1 :
                        data_out.append(8)
                        break
                    elif 4.1>score>=3.9 :
                        data_out.append(6)
                        break
                    else :
                        data_out.append(4)
                        break
                else:
                    data_out.append(0)
                    break
                        
            if data[i] == '소재지':
                b = data[i+5].find('>')
                data_out.append(data[i+5][:b])
                break
                
            if data[i] ==df[j]:
                data_out.append(data[i+5])
    return data_out

# TOEIC 점수 - 900이상 10점 , 900~ 800 8점 , 800 ~ 700 6점, 700 이하 4점     
def TOEIC(data):
    data_out = []
    for i in range(len(data)):
        if data[i] == 'TOEIC':
            toeic_score = data[i+5]
    if toeic_score != '':
        toeic_score = int(toeic_score.replace('점', ''))

        if toeic_score >900 :
            data_out.append(10)
        elif 900 >= toeic_score >800 : 
            data_out.append(8)
        elif 800 >= toeic_score >700 : 
            data_out.append(6)
        else : 
            data_out.append(4)
    elif toeic_score == '':
        data_out.append(0)
    return data_out

# 자격증 - 기사 자격증 4점 , 그외 기타 자격증 1점 (가산점)
def certificate(data):
    tot = []
    data_out = []
    i = data.index('점수')
    while True:
        i += 1
        if data[i] != '기간':
            tot.append(data[i])
        else: 
            break
    del tot[3]
    list_1 = []
    for i in range(0,len(tot),3):
        list_1.append(tot[i])
    score = 0
    for i in range(len(list_1)) :
        if len(list_1) <= 0:
            break 
        else :
            if list_1[i].find('기사') == 4:
                 score += 5
            else : 
                score += 1
    data_out.append(score)
    return data_out

# 교육/활동/수상 - 1회당 2점 (가산점)
def Education(data):
    data_out = []
    i = data.index('군별')
    j = data.index('계급')
    k = data.index('복무기간')
    tot = []
    if data[i-1] != '':
        tot.append(data[i-1])
        if data[j-1] != '':
            tot.append(data[j-1])
            if data[k-1] != '':
                tot.append(data[k-1])
    data_out.append(len(tot))
    return data_out

# 군필여부, 군별, 계급 , 복무기간
def military(data):
    data_out = []
    df3 = ['군필여부','군별','계급','복무기간']
    for i in range(len(data)):
        for j in range(len(df3)):
            if data[i] == '복무기간':
                a = data[i+1].find('>')
                data_out.append(data[i+1][:a])
                break
            if data[i] == df3[j]:
                data_out.append(data[i+1])
    return data_out

# 컴퓨터 언어 능력 - 상: 2점 중: 1점 하: 0점 ///최대 10점 
def Language_ability(data):
    data_out = []
    df = [ 'R','Python','Javascript','SQL','HTML/CSS']
    test = []
    for i in range(len(data)):
        for j in range(len(df)):
            if data[i] == 'HTML/CSS':
                a = data[i+5].find('>')
                test.append(data[i+4][:a])
                break
            if data[i] == df[j]:
                test.append(data[i+5])
    score = 0
    for i in test :
        if i == '상' :
            score +=2
        elif i == '중' :
            score += 1
        else : 
            score += 0
    data_out.append(score)

    return data_out

# 종합 !!
def total(data):
    data_out = []
    data = data.split('><')
    data_out.extend(Privacy(data))
    data_out.extend(School(data))
    data_out.extend(TOEIC(data))
    data_out.extend(certificate(data))
    data_out.extend(Education (data))
    data_out.extend(military(data))
    data_out.extend(Language_ability(data))

    return data_out
```
